```python
import random
List = ["python", "java", "javascript", "computer"]
computer_choice = random.choice(List)
if computer_choice in List:
    computer = list(computer_choice)
    List2 = []
    for i in range(len(computer)):
       
        blank = "_"
        List2.append(blank)
    print(" ".join(List2))
    for i in range(len(computer)):
       user_choice = input("Enter your choice:")
       if user_choice in computer:
           for j in range(len(computer)):
               if computer[j] == user_choice:
                   List2[j] = user_choice
       else:
           print("wrong guess") 
            
       print(" ".join(List2))
       
print(f"the word was {computer_choice}")
        
```
