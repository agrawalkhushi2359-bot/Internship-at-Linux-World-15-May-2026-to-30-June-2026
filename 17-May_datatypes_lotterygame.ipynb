```python
#dictionary
d={
    "name":"khushi agrawal",
    "age":21
}
print(d)
```

```python
#dictionary functions
print(d.keys())
print(d.values())
print(d.get("name"))
print(d.copy())
print(d.items())
print(d.setdefault("name","Agrawal"))
d.update({"age":50})
print(d)
print(d.items())
```

```python
#tuple= immutable,unique value only,allow duplicate values
t=("khushi agrawal",200,50.12,200)
print(t.index(200))
```

```python
#set=immutable, non repeatable value
s={1,2,3}
print(s)
s.add(4)
print(s)
print(s.copy())
p={2,3,4,5,6,7,8}
print(s.difference(p))
print(s.intersection(p))
print(s.union(p))
print(s.update(p))
print(s.symmetric_difference(p))
```
